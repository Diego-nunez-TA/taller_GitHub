```python
table_1 = soup_1.find_all('table')[1]
```

```python
def extraer_nombres(table):
    nombres_1 = []
    for row in table_1.find_all('tr'):
        celdas = row.find_all('td')
        if len(celdas) > 1:
            link = celdas[1].find('a')
            nombres_1.append(link.get_text())
    return nombres_1
```
